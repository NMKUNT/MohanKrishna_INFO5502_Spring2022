```python
# Your answer here (python code + explanation):

import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt 
import random
import pandas as pan
import matplotlib


df = pan.read_csv ('/content/lab_assignment_05_census.csv')
for i in df.index:  # replacing 1 with male and 2 with female in our dataset
  if df['Sex'][i]==1:
     df['Sex'][i]='Male'
  else:
    df['Sex'][i]='Female'

sns.set(rc={'figure.figsize':(25,10)}) # settign the output size of the scatterplot
plt.ticklabel_format(style='plain', axis='y') # To print the complete number on y axis instead of e^7
ax = sns.scatterplot(x="Age", y="People", data=df, hue = 'Sex', size = 'Year') # this plots our graph
plt.ylabel("Number of People") # renaming Y lable
```

We have collected all the articles from AMIA in the last 30 years. The dataset is avaliable on GitHub: https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv. 

Please use the tool Yake: https://github.com/LIAAD/yake to extract the top N (you can set different N value to get the best visualization results) keywords from the article titles (all the articles), then visualize how the keywords changed over years in one figure. (Split the data by every five years)

```python
import pandas as pd
import matplotlib.pyplot as plt
import yake
import random
import matplotlib

amia=pd.read_csv("https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv")#reads the file
amia=amia[amia['year'].isna()!=True] 
amia['year']=amia['year'].astype('int64')
```

```python
#This code prints the top words of each year
wordcount=random.randint(5,20) # choose on random integer between 5 and 20, which is going to be our top number of keywords

for year in amia['year'].unique(): # selects unique years 
  print("Top ",wordcount," keywords of the year {} are: ".format(year)) 
  wordslist=" ".join(title for title in amia[amia['year']==year]['title'].values) # creating the list of words 
  extract=yake.KeywordExtractor(lan="en", top=wordcount, features=None) #yake helps to extracts the keywords
  words=extract.extract_keywords(wordslist)
  Score = [y for x, y in words]
  Keyword=[x for x, y in words] 			
  for a in words:
    print(a)

```

```python
# this below code prints the top 5 words for each 5 year sets

wordcount = 5
j=0
while(j<9):
	i=0 + (5*j)
	for y in range(1981+i,1986+i): # selects unique years 
		wordslist=" ".join(title for title in amia[amia['year']==y]['title'].values) # creating the list of words 
	extract=yake.KeywordExtractor(lan="en", top=wordcount, features=None) #yake helps to extracts the keywords
	words=extract.extract_keywords(wordslist)
	Score = [y for x, y in words]
	Keyword=[x for x, y in words]
	plt.scatter(Keyword,Score,label=y)
	plt.title("Keyword Trend")
	j = j+1
	print("from year ",1981+i," to ",1986+i)
	for a in words:
		print(a)
	print("-------------------")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=10)
matplotlib.rc('font', size=5)
plt.xticks(rotation=90)
plt.xlabel("Keyword")
plt.ylabel("Score")
plt.show()
j = j+1
```
